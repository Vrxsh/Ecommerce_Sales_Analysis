```python
import pandas as pd
import random

# -----------------------------
# Dataset Creation
# -----------------------------
random.seed(42)

regions = ['North', 'South', 'East', 'West']
categories = ['Electronics', 'Clothing', 'Home & Garden', 'Sports', 'Books']
salespersons = ['Alice', 'Bob', 'Carol', 'David', 'Emma', 'Frank']

data = {
    'transaction_id': range(1001, 1201),
    'region': [random.choice(regions) for _ in range(200)],
    'category': [random.choice(categories) for _ in range(200)],
    'salesperson': [random.choice(salespersons) for _ in range(200)],
    'sales_amount': [round(random.uniform(50, 5000), 2) for _ in range(200)],
    'customer_id': [random.randint(5000, 5100) for _ in range(200)]
}

df = pd.DataFrame(data)
print(df.head())
print(f"\nDataset shape: {df.shape}")
```

Task 1: Basic Grouping and Single Aggregations

```python
# 1. Total sales per region
region_sales = df.groupby('region')['sales_amount'].sum().reset_index()
region_sales_sorted = region_sales.sort_values('sales_amount', ascending=False)
print("\nTotal Sales by Region:\n", region_sales_sorted)

# 2. Number of transactions per category
category_count = df.groupby('category')['transaction_id'].count().reset_index()
print("\nTransaction Count by Category:\n", category_count)

# 3. Average sales per salesperson
salesperson_avg = df.groupby('salesperson')['sales_amount'].mean().reset_index()
print("\nAverage Sales by Salesperson:\n", salesperson_avg)
```

Task 2: Multi-Column Grouping and Multiple Aggregations

```python
# 1. Region + Category total sales
region_category_sales = df.groupby(['region','category'])['sales_amount'].sum().reset_index()
print("\nSales by Region & Category:\n", region_category_sales)

# 2. Salesperson metrics
salesperson_metrics = df.groupby('salesperson')['sales_amount'].agg(['sum','mean','count']).reset_index()
salesperson_metrics_sorted = salesperson_metrics.sort_values('sum', ascending=False)
print("\nSalesperson Performance:\n", salesperson_metrics_sorted)

# 3. Top category by revenue
top_category = df.groupby('category')['sales_amount'].sum().idxmax()
print("\nCategory with Maximum Revenue:", top_category)
```

• 	Top Region: From Task 1, the region with the highest sales is identified via sorting.
• 	Top Salesperson: From Task 2, sorting by  shows the best performer.
• 	Top Category:  reveals the category with maximum revenue.
• 	Custom Aggregation: Sales range highlights variability in transaction values per region.
• 	Final Report: Multi-level columns show combined metrics (total sales, average, transaction count).
